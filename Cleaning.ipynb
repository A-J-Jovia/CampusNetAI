```python
import pandas as pd
import numpy as np

np.random.seed(42)

# ==================================================
# CAMPUS CONFIGURATION
# ==================================================

locations = [
    ("MAIN_BLOCK", "GROUND", "Library"),
    ("MAIN_BLOCK", "GROUND", "Chemistry Lab"),
    ("MAIN_BLOCK", "GROUND", "Classroom"),
    ("MAIN_BLOCK", "FLOOR1", "Auditorium"),
    ("MAIN_BLOCK", "FLOOR1", "Classroom"),
    ("MAIN_BLOCK", "FLOOR2", "Code Studio"),
    ("MAIN_BLOCK", "FLOOR2", "IT Project Lab"),
    ("MAIN_BLOCK", "FLOOR2", "Computer Lab"),
    ("MAIN_BLOCK", "FLOOR3", "Cybernet Lab"),
    ("MAIN_BLOCK", "FLOOR3", "Classroom"),

    ("MECH_BLOCK", "GROUND", "FABLAB"),
    ("MECH_BLOCK", "GROUND", "CAD Lab"),
    ("MECH_BLOCK", "GROUND", "Lab"),
    ("MECH_BLOCK", "FLOOR1", "Classroom"),
    ("MECH_BLOCK", "FLOOR2", "Library"),
    ("MECH_BLOCK", "FLOOR2", "Lab"),
    ("MECH_BLOCK", "FLOOR3", "Lab"),

    ("AI_BLOCK", "GROUND", "Robospace"),
    ("AI_BLOCK", "GROUND", "AI Lab"),
    ("AI_BLOCK", "FLOOR1", "ML Lab"),
    ("AI_BLOCK", "FLOOR1", "Classroom"),
    ("AI_BLOCK", "FLOOR2", "GenAI Lab"),
    ("AI_BLOCK", "FLOOR2", "Synapse Studio"),
    ("AI_BLOCK", "FLOOR2", "Vista Hall"),
    ("AI_BLOCK", "FLOOR2", "Classroom"),
    ("AI_BLOCK", "FLOOR3", "Cyber Lab"),
    ("AI_BLOCK", "FLOOR3", "Data Visualization Lab"),

    ("AMENITY", "GROUND", "Food Court"),
    ("AMENITY", "GROUND", "Music Room"),

    ("MARIO", "GROUND", "Snack Corner"),

    ("HOSTEL_BOYS", "GROUND", "Hostel"),
    ("HOSTEL_GIRLS", "GROUND", "Hostel"),

    ("BUS_AREA", "GROUND", "Bus Bay")
]

events = [
    "No_Event",
    "No_Event",
    "No_Event",
    "No_Event",
    "No_Event",
    "Thiran",
    "Freshwarite",
    "Culturals",
    "FlashMob"
]

weather_conditions = [
    "Sunny",
    "Cloudy",
    "Rainy"
]

rows = []

# ==================================================
# GENERATE 10,000 RECORDS
# ==================================================

for i in range(10000):

    building, floor, zone = locations[
        np.random.randint(0, len(locations))
    ]

    timestamp = pd.Timestamp("2025-01-01") + pd.to_timedelta(
        np.random.randint(0, 365*24*60),
        unit="m"
    )

    hour = timestamp.hour

    weather = np.random.choice(weather_conditions)

    event = np.random.choice(events)

    # ==================================================
    # BASE USERS
    # ==================================================

    users = np.random.randint(40, 120)

    # Academic hours

    if building in ["MAIN_BLOCK", "AI_BLOCK", "MECH_BLOCK"]:

        if 9 <= hour <= 12:
            users += np.random.randint(50, 120)

        elif 13 <= hour <= 16:
            users += np.random.randint(30, 90)

    # Lunch rush

    if building in ["AMENITY", "MARIO"]:

        if 12 <= hour <= 14:
            users += np.random.randint(80, 180)

    # Hostel rush

    if building in ["HOSTEL_BOYS", "HOSTEL_GIRLS"]:

        if 18 <= hour <= 23:
            users += np.random.randint(100, 250)

    # Events

    if event != "No_Event":

        if zone in [
            "Auditorium",
            "Vista Hall",
            "Food Court",
            "Synapse Studio"
        ]:
            users += np.random.randint(100, 250)

    # ==================================================
    # NETWORK METRICS
    # ==================================================

    bandwidth_usage = round(
        users * np.random.uniform(6, 15),
        2
    )

    download_speed = round(
        max(8, 150 - users * 0.55),
        2
    )

    upload_speed = round(
        max(5, 75 - users * 0.25),
        2
    )

    latency = round(
        8 + users * 0.65,
        2
    )

    packet_loss = round(
        min(
            15,
            np.random.uniform(0.5, 2.5)
            + users * 0.02
        ),
        2
    )

    signal_strength = np.random.randint(
        -80,
        -35
    )

    # ==================================================
    # TARGET
    # ==================================================

    if users < 100:
        congestion = "Low"

    elif users < 180:
        congestion = "Medium"

    else:
        congestion = "High"

    rows.append([
        i+1,
        timestamp,
        building,
        floor,
        zone,
        users,
        bandwidth_usage,
        download_speed,
        upload_speed,
        latency,
        packet_loss,
        signal_strength,
        weather,
        event,
        congestion
    ])

# ==================================================
# DATAFRAME
# ==================================================

columns = [
    "Record_ID",
    "Timestamp",
    "Building",
    "Floor",
    "Zone",
    "Connected_Users",
    "Bandwidth_Usage_MB",
    "Download_Speed_Mbps",
    "Upload_Speed_Mbps",
    "Latency_ms",
    "Packet_Loss_Percent",
    "Signal_Strength_dBm",
    "Weather",
    "Event_Occurring",
    "Congestion_Status"
]

df = pd.DataFrame(rows, columns=columns)

# ==================================================
# ADD MISSING VALUES
# ==================================================

download_missing = np.random.choice(
    df.index,
    size=300,
    replace=False
)

df.loc[
    download_missing,
    "Download_Speed_Mbps"
] = np.nan

signal_missing = np.random.choice(
    df.index,
    size=200,
    replace=False
)

df.loc[
    signal_missing,
    "Signal_Strength_dBm"
] = np.nan

# ==================================================
# ADD DUPLICATES
# ==================================================

duplicates = df.sample(
    200,
    random_state=42
)

df = pd.concat(
    [df, duplicates],
    ignore_index=True
)

# ==================================================
# SAVE
# ==================================================

df.to_csv(
    "campus_wifi_logs.csv",
    index=False
)

print("="*60)
print("DATASET GENERATED SUCCESSFULLY")
print("="*60)

print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")

print("\nCongestion Distribution:")
print(df["Congestion_Status"].value_counts())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())

print("\nFirst 5 Rows:")
display(df.head())
```

# Section 2: Data Cleaning and Preprocessing

In this section, the dataset is cleaned to improve data quality.

The following preprocessing steps are performed:

1. Handling missing values
2. Removing duplicate records
3. Verifying data types
4. Creating a cleaned dataset
5. Saving the cleaned dataset as a new CSV file

This cleaned dataset will be used in subsequent SQL, ETL, PySpark, and Machine Learning tasks.

```python
df_clean = df.copy()

print("Dataset copied successfully")
print(f"Shape: {df_clean.shape}")
```

```python
print("="*60)
print("MISSING VALUES BEFORE CLEANING")
print("="*60)

display(df_clean.isnull().sum())
```

```python
df_clean["Download_Speed_Mbps"] = df_clean[
    "Download_Speed_Mbps"
].fillna(
    df_clean["Download_Speed_Mbps"].median()
)

print("Download Speed Missing Values Handled")
```

```python
df_clean["Signal_Strength_dBm"] = df_clean[
    "Signal_Strength_dBm"
].fillna(
    df_clean["Signal_Strength_dBm"].median()
)

print("Signal Strength Missing Values Handled")
```

```python
print("="*60)
print("MISSING VALUES AFTER CLEANING")
print("="*60)

display(df_clean.isnull().sum())
```

```python
print("="*60)
print("DUPLICATE RECORDS BEFORE CLEANING")
print("="*60)

duplicates_before = df_clean.duplicated().sum()

print(f"Duplicate Rows: {duplicates_before}")
```

```python
df_clean.drop_duplicates(inplace=True)
print("Duplicate Records Removed")
```

```python
print("="*60)
print("DUPLICATE RECORDS AFTER CLEANING")
print("="*60)

duplicates_after = df_clean.duplicated().sum()

print(f"Duplicate Rows: {duplicates_after}")
```

```python
print("="*60)
print("DATA TYPES")
print("="*60)

display(df_clean.dtypes)
```

```python
df_clean["Timestamp"] = pd.to_datetime(
    df_clean["Timestamp"]
)

print("Timestamp converted to datetime")
```
